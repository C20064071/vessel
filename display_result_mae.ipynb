```python
import numpy as np
import pandas as pd

# path='/root/work/result/mae_k'
# path='/root/work/result/mae_mlp_k'
# path='/root/work/result/mae_schedule'
# path='/root/work/result/MAE/MLP'
# path='/root/work/result/MAE/decls'
# path='/root/work/result/MAE/decls_sche'
# path='/root/work/result/MIM/basic'
# path='/root/work/result/MIM/mask50'
# path='/root/work/result/MIM/mask25'
path='/root/work/result/RMSPE/TMD'

# Load the contents of the file
data = np.load(path+'/log.npy')

# Display the contents
print(data.shape)
```

```python
result = []
for i in range(5):
    min_index = np.argmin(data[i][:,2])
    result.append(data[i][min_index])


df_result = pd.DataFrame(result)

print(df_result)
df_result.to_csv(path+'/result.csv', index=False)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 5, figsize=(25, 5))

for i in range(5):
    axs[i].plot(data[i][:, 0], data[i][:, 1], label='train_fom')
    axs[i].plot(data[i][:, 0], data[i][:, 2], label='val_fom')
    axs[i].set_xlabel('epoch')
    axs[i].set_ylabel('Loss')
    axs[i].set_ylim(0, 0.2)
    axs[i].legend()

plt.tight_layout()
plt.show()
fig.savefig(path+'/fom_graph.png')
```
